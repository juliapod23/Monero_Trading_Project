# Monero Feature Visuals

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

feature_path = "../data/processed/xmr_features.csv"
assert os.path.exists(feature_path), "Feature CSV not found! Run feature_engineering.py first."

df = pd.read_csv("../data/processed/xmr_features.csv", parse_dates=["time"])
print("Feature data loaded successfully:")
df.head()

!python ../scripts/feature_engineering.py
```

```python
# plot log returns
plt.figure(figsize=(12, 4))
plt.plot(df["time"], df["log_return"])
plt.title("XMR Log Returns")
plt.xlabel("Date")
plt.ylabel("Log Return")
plt.grid()
plt.show()
```

```python
# plot volatility
plt.figure(figsize=(12, 4))
plt.plot(df["time"], df["volatility_14"], color="orange")
plt.title("14-Day Rolling Volatility")
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.grid()
plt.show()
```

```python
# plot rsi
plt.figure(figsize=(12, 4))
plt.plot(df["time"], df["rsi_14"], color="green")
plt.axhline(70, color='red', linestyle='--')
plt.axhline(30, color='blue', linestyle='--')
plt.title("RSI (14)")
plt.xlabel("Date")
plt.ylabel("RSI")
plt.grid()
plt.show()
```

```python
# plot z-score
plt.figure(figsize=(12, 4))
plt.plot(df["time"], df["zscore_14"], color="purple")
plt.axhline(2, color='red', linestyle='--')
plt.axhline(-2, color='blue', linestyle='--')
plt.title("Z-score of Closing Price (14-day)")
plt.xlabel("Date")
plt.ylabel("Z-score")
plt.grid()
plt.show()
```
